# 单变量、双变量以及多变量分析
## seaborn中的iris数据实现分析操作

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.tail() #最后五行数据
```

```python
df.shape
```

```python
df["species"].unique() # 查看species中有多少中数据
```

```python
df.loc[df["species"] == "virginica"]  #显示virginica的所有数据
```

```python
df_virginica = df.loc[df["species"]=="virginica"]
df_setosa = df.loc[df["species"]=="setosa"]
df_versicolor = df.loc[df["species"]=="versicolor"]
```

## 单变量分析

```python
plt.plot(df_setosa["sepal_length"],marker='x')
plt.plot(df_versicolor["sepal_length"],marker='*')
plt.plot(df_virginica["sepal_length"],marker='o')
plt.ylabel("sepal_length")
plt.show()
#绘制
```

```python
plt.plot(df_setosa["sepal_width"],marker='x')
plt.plot(df_versicolor["sepal_width"],marker='*')
plt.plot(df_virginica["sepal_width"],marker='o')
plt.ylabel("sepal_width")
plt.show()
```

```python
plt.plot(df_setosa["petal_length"],marker='x')
plt.plot(df_versicolor["petal_length"],marker='*')
plt.plot(df_virginica["petal_length"],marker='o')
plt.ylabel("petal_length")
plt.show()
```

```python
plt.plot(df_setosa["petal_width"],marker='x')
plt.plot(df_versicolor["petal_width"],marker='*')
plt.plot(df_virginica["petal_width"],marker='o')
plt.ylabel("petal_width")
plt.show()
```

```python
plt.plot(df_setosa["species"],df_setosa["petal_width"],marker='x')
plt.plot(df_versicolor["species"],df_versicolor["petal_width"],marker='*')
plt.plot(df_virginica["species"],df_virginica["petal_width"],marker='o')
plt.ylabel("petal_width")
plt.xlabel("species")
plt.show()
```

```python
plt.plot(df_setosa["petal_length"],np.zeros_like(df_setosa["petal_length"]),marker='x')
plt.plot(df_versicolor["petal_length"],np.zeros_like(df_versicolor["petal_length"]),marker='*')
plt.plot(df_virginica["petal_length"],np.zeros_like(df_virginica["petal_length"]),marker='o')
plt.ylabel("petal_length")
plt.show()

#使用np.zeros_like() 将所有的值放在0值上
```

# 双变量分析

```python
plt.scatter(df_setosa["petal_length"],df_setosa["petal_width"],marker='x')
plt.scatter(df_versicolor["petal_length"],df_versicolor["petal_width"],marker='*')
plt.scatter(df_virginica["petal_length"],df_virginica["petal_width"],marker='o')

plt.ylabel("petal_width")
plt.xlabel("petal_length")
plt.show
```

```python
sns.FacetGrid(df, hue="species", height=7).map(plt.scatter,"petal_length","petal_width").add_legend()
```

# 多变量分析

```python
sns.pairplot(df,hue="species",height=4)
```

# 柱状图

```python
sns.distplot(df_setosa["sepal_length"])
sns.distplot(df_versicolor["sepal_length"])
sns.distplot(df_virginica["sepal_length"])
```

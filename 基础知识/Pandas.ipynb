```python
import pandas as pd
import numpy as np
```

```python
arr_1 = np.arange(0,20).reshape(5,4)
```

```python
pd.DataFrame(arr_1) #DataFrame 将矩阵变为一个表结构
```

```python
df = pd.DataFrame(arr_1,index=['row1','row2','row3','row4','row5'],columns = ['col1','col2','col3','col4'])
#使用index设置行名，使用columns设置列名
```

```python
df
```

```python
a = df.loc['row1']  #使用loc进行索引，需要指定行（确定元素也可以添加上列）  df.loc[hang,lie]
b = df.loc['row1','col1']
print(b)
```

```python
df.loc[:,'col1']  #采用这种方法可以显示列
```

```python
df.iloc[:,1:].values   #values可显示数组
```

```python
df.iloc[:,]
```

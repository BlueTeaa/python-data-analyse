```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('data/test1.xlsx') 
```

```python
df.head(5)
```

```python
df_Achnanthes_flexella = df.loc[df["Species"]=="Achnanthes flexella"]
df_Achnanthes_spp = df.loc[df["Species"]=="Achnanthes spp."]
df_Achnanthes_ehrenbergii = df.loc[df["Species"]=="Actinocyclus ehrenbergii"]
```

```python
plt.plot(df_Achnanthes_flexella["5cm"],marker='x')
plt.plot(df_Achnanthes_spp["5cm"],marker='*')
plt.plot(df_Achnanthes_ehrenbergii["5cm"],marker='o')
plt.ylabel("sepal_length")
plt.show()
#绘制
```

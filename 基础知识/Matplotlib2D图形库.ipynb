# MatplotLib   2D绘图库

```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示正负号

%matplotlib inline #作用：后续显示图像不需要每次都写plt.show()
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g') #scatter 表示使用散点图的形式显示
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(a,b,c='r')
plt.title('2D图表实例')
plt.xlabel('x轴')
plt.ylabel('y轴')

plt.savefig('Test.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'r*',linestyle='solid',linewidth=1,markersize=8)
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.title('2D Diagram')
```

```python
plt.subplot(2,2,1)  # subplot 表示几行几列 Eg:两行两列，第一个图像
plt.plot(x,y,'r-')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g:')
plt.subplot(2,2,4)
plt.plot(x,y,'y-')
```

```python
np.pi
```

```python
x = np.arange(0, 4*np.pi, 0.1)  #0-4pi 步长为0.1
y = np.sin(x)

plt.plot(x,y,'g-')
```

## 条形图 plt.bar

```python
x = [2,8,10]
y = [11,16,9]

x2 = [5,9,7]
y2 = [16,4,21]

plt.bar(x,y)
plt.bar(x2,y2,color='g')

plt.ylabel('Y')
plt.xlabel('X')
plt.title('Bar graph')
```

## Histograms 直方图

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27,58,41,21,20])
plt.hist(a,bins=40)
plt.title('Histograms')
```

## Box plot using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert='True')

#图中产生的圆圈表示偏离值
```

## Pie Chart

```python
#Data
labels = 'python', 'C++', 'Javascript', 'Go'
sizes = [215,135,245,210]
color = ['yellow','red','blue','green']
explode=(0.1,0.1,0,0)#表示饼之间的分离度


plt.pie(sizes,explode=explode,labels=labels,colors=color, autopct='%1.2f%%',shadow='Ture')
# autopct表示保留的小数位 shadow表示是否有阴影
```
